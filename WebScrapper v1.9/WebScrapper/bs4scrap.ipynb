```python
from bs4 import BeautifulSoup
import requests
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/97.0.4692.71 Safari/537.36'}
import pandas as pd
import Getlink
from urllib.error import HTTPError
```

```python
products=[]
prices=[]
links=[]
images=[]
print(Getlink.get_amz_url('Iphones'))
source=requests.get(Getlink.get_amz_url('Iphones'),headers=headers)
source=source.content
#print(source)
soup = BeautifulSoup(source,'lxml')
search_results= soup.find('div',{'class':'s-main-slot'})
soup
```

```python
i=0
count=0
try:
   while count<5:
      result_row=search_results.find('div',{'data-index':i},{'data-cel-widget':'search_result_'+chr(i)})
      print(result_row)
      
      if not result_row['data-asin'] or 'AdHolder' in result_row['class']:
         i+=1
         
      else:       
         products.append(result_row.find('span',{"class":"a-size-medium a-color-base a-text-normal"}).text)
         prices.append(result_row.find('span', {'class': 'a-price-whole'}).text)
         links.append('https://www.amazon.in'+result_row.find('a',{'class':'a-link-normal a-text-normal'},{'target':'_blank'})['href'])
         images.append(result_row.find('img',{'class':'s-image'})['src'])
         i+=1
         count+=1
except Exception as err:
   print( err)
```

```python
lists = [products, prices,links,images]
df = pd.concat([pd.Series(x) for x in lists], axis=1)
df  
```
