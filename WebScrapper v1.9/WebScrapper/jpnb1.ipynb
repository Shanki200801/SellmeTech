```python
param=['$400,00.98','28,999']
par=[param]
def remove_char(x):
    import re
    import locale
    decimal_point_char = locale.localeconv()['decimal_point']
    clean = re.sub(r'[^0-9'+decimal_point_char+r']+', '', str(x))
    return float(clean)

#'{:,}'.format(float(clean))


par[0] = [ remove_char(i) for i in par[0]]  
par
```
